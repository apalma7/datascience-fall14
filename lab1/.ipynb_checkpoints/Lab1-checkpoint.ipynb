```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()

diabetes_X_train = diabetes.data[:-20]
diabetes_X_test = diabetes.data[-20:]

diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

regr = LinearRegression()

regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
print "Coefficients:", regr.coef_
print("Residual sum of squares: %.2f" % np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))
```

```python
from sklearn.feature_selection import RFE

rfe = RFE(estimator=regr, n_features_to_select=2, step=1)
rfe.fit(diabetes_X_train, diabetes_y_train)
print rfe.ranking_
```

```python
#The total number of records in a feature
total_data = boston.data.shape[0]
#The values belowe represent the P% of the data (for instance boston_20_X represents 20% of the data).
#Each value has a corresponding compliment value such that together they reprsent 100% of the data.
#For example, boston_20_X reprsents the top 20% of the data and boston_80_X represents the bottom 80%.
boston_20_X = boston.data[:math.floor(.2*total_data)]
boston_80_X = boston.data[math.ceil(.2*total_data):]
boston_40_X = boston.data[:math.floor(.4*total_data)]
boston_60_X = boston.data[math.ceil(.4*total_data):]

boston_20_y = boston.target[:math.floor(.2*total_data)]
boston_80_y = boston.target[math.ceil(.2*total_data):]
boston_40_y = boston.target[:math.floor(.4*total_data)]
boston_60_y = boston.target[math.ceil(.4*total_data):]


regr_20 = LinearRegression()
regr_40 = LinearRegression()
regr_60 = LinearRegression()
regr_80 = LinearRegression()

regr_20.fit(boston_20_X, boston_20_y)
regr_40.fit(boston_40_X, boston_40_y)
regr_60.fit(boston_60_X, boston_60_y)
regr_80.fit(boston_80_X, boston_80_y)

predicted_20 = regr_20.predict(boston_80_X)
predicted_40 = regr_40.predict(boston_60_X)
predicted_60 = regr_60.predict(boston_40_X)
predicted_80 = regr_80.predict(boston_20_X)

mse_func = lambda predicted, y: np.mean((predicted - y) ** 2)
mses = [mse_func(predicted_20, boston_80_y), mse_func(predicted_40, boston_60_y), mse_func(predicted_60, boston_40_y), mse_func(predicted_80, boston_20_y)]
barx = [20, 40, 60, 80]

plt.bar(barx, mses, width=5, align='center')
plt.show()
```

```python
rfes = []
regrs = [regr_20, regr_40, regr_60, regr_80]
xy = [ (boston_20_X, boston_20_y), (boston_40_X, boston_40_y), (boston_60_X, boston_60_y), (boston_80_X, boston_80_y) ]

for i in range(0, 4):
    rfes.append(RFE(estimator=regrs[i], n_features_to_select=2, step=1))
    rfes[i].fit(xy[i][0], xy[i][1])
    print rfes[i].ranking_

print boston.feature_names
```

```python
from sklearn.datasets import make_classification, make_moons

X_temp, y_temp = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=1, n_classes=2)
X, y = make_moons(noise=0.3, random_state=0)

colormap = np.array(['r', 'g'])
plt.scatter(X[:,0], X[:,1], marker='o', c=colormap[y], s=50)
plt.show()
```

```python
X_train = X[:-20,:]
X_test = X[-20:,:]
y_train = y[:-20]
y_test = y[-20:]
```

```python
from sklearn import svm

clf_linear = svm.SVC(kernel='linear')
clf_linear.fit(X_train, y_train)

pred = clf_linear.predict(X_test)

errors = 0.0
for i in range(len(pred)):
    if pred[i] != y_test[i]:
              errors += 1.0
print "Test data errors = ", errors, "out of ", len(pred), " predictions"

total_pred = clf_linear.predict(X)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != y[i]:
        errors += 1.0
print "Total errors = ", errors, " out of ", len(total_pred), " predictions"
```

```python
h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = clf_linear.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
pylab.contourf(xx, yy, Z, cmap=pylab.cm.Paired)
pylab.axis('off')

pylab.scatter(X[:, 0], X[:, 1], c=colormap[y], s=50, cmap=pylab.cm.Paired)
```

```python
clf_non_linear = svm.SVC(kernel='rbf')
clf_non_linear.fit(X_train, y_train)

pred = clf_non_linear.predict(X_test)
errors = 0.0
for i in range(len(pred)):
    if pred[i] != y_test[i]:
        errors += 1.0
print "Test data errors =", errors, "out of", len(pred), "predictions"

total_pred = clf_non_linear.predict(X)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != y[i]:
        errors += 1.0
print "Total errors =", errors, "out of", len(total_pred), "predictions"

h = .02
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = clf_non_linear.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
pylab.contourf(xx, yy, Z, cmap=pylab.cm.Paired)
pylab.axis("off")

pylab.scatter(X[:, 0], X[:, 1], c=colormap[y], s=50, cmap=pylab.cm.Paired)
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

total_iris_data = iris.data.shape[0]

X_train = iris.data[:math.floor(.8*total_iris_data),]
X_test = iris.data[math.ceil(.8*total_iris_data):,]
y_train = iris.target[:math.floor(.8*total_iris_data),]
y_test = iris.target[math.ceil(.8*total_iris_data):,]

#Linear
clf_linear = svm.SVC(kernel='linear')
clf_linear.fit(X_train, y_train)

pred = clf_linear.predict(X_test)

errors = 0.0
for i in range(len(pred)):
    if pred[i] != y_test[i]:
              errors += 1.0
print "Test data errors (linear) =", errors, "out of", len(pred), "predictions"

total_pred = clf_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != iris.target[i]:
        errors += 1.0
print "Total errors (linear) =", errors, "out of", len(total_pred), "predictions"

#Non-linear
clf_non_linear = svm.SVC(kernel='rbf')
clf_non_linear.fit(X_train, y_train)

pred = clf_non_linear.predict(X_test)
errors = 0.0
for i in range(len(pred)):
    if pred[i] != y_test[i]:
        errors += 1.0
print "Test data errors (non-linear) =", errors, "out of", len(pred), "predictions"

total_pred = clf_non_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != iris.target[i]:
        errors += 1.0
print "Total errors (non-linear) =", errors, "out of", len(total_pred), "predictions"
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=5, n_classes=2)

plt.scatter(X[:, 0], X[:, 1], marker='o', c=colormap[y], s=50)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics

kmeans = KMeans(init='random', n_clusters=2, n_init=10)
kmeans.fit(X)

print "Homogeneity_score =", metrics.homogeneity_score(y, kmeans.labels_)

print "Completeness_score =", metrics.completeness_score(y, kmeans.labels_)
```

```python
h = .02
x_min, x_max = -2, 4
y_min, y_max = -2, 4
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)

plt.clf()
plt.imshow(Z, interpolation='nearest',cmap=plt.cm.Paired, aspect='auto', origin='lower',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()))

plt.scatter(X[:, 0], X[:,1], marker='o', c=colormap[y], s=50)
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=169, linewidths=3, color='w', zorder=10)

plt.title('K-means cluster on synthetic data\nCentroids are marked with a white cross')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
